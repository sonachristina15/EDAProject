#### Chart - 14 - Pair Plot

````python
# Pair Plot visualization code

import seaborn as sns
import matplotlib.pyplot as plt

# Select important numerical variables
cols = ['Annual_Income',
        'Outstanding_Debt',
        'Credit_Utilization_Ratio',
        'Monthly_Balance',
        'Interest_Rate',
        'Credit_Score']

# Remove missing values (recommended)
pair_df = df[cols].dropna()

# Pair Plot
sns.pairplot(
    pair_df,
    hue='Credit_Score',
    diag_kind='hist',
    corner=True,
    plot_kws={'alpha': 0.6}
)

plt.suptitle('Pair Plot of Financial Variables by Credit Score', y=1.02)
plt.show()
````
